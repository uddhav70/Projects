```python
import pandas as pd
data_crime2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Crime_prediction/crime_all_2022.csv')
data_crime2
```

```python
# Now In the data the states are converted to numerical for ease of plotting the data and avoid overlapping of X-axix in colab and clustering.
```

```python
data_state = data_crime2

STATES = {'andhra pradesh':1, 'arunachal pradesh':2, 'assam':3, 'bihar':4, 'chhattisgarh':5, 'goa':6, 'gujarat':7, 'haryana':8,'himachal pradesh':9,'jammu & kashmir':10,'jharkhand':11,'karnataka':12, 'kerala':13, 'madhya pradesh':14, 'maharashtra':15, 'manipur':16, 'meghalaya':17,
'mizoram':18, 'nagaland':19, 'odisha':20, 'punjab':21, 'rajasthan':22, 'sikkim':23, 'tamil nadu':24,
 'tripura':25, 'uttar pradesh':26, 'uttarakhand':27, 'west bengal':28, 'a & n islands':29,
'chandigarh':30, 'd & n haveli':31, 'daman & diu':32,'lakshadweep':33,'puducherry':34
}
  
data_state.STATES = [STATES[item] for item in data_state.STATES]
data_state
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

plt.scatter(x =data_state.STATES , y =data_state.MURDER , color = 'green')
plt.xlabel('STATE')
plt.ylabel('MURDER')
plt.show()
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
for p in ['MURDER',	'ATTEMPT TO MURDER', 'CULPABLE HOMICIDE NOT AMOUNTING TO MURDER', 'RAPE', 'OTHER RAPE',	'KIDNAPPING & ABDUCTION', 'KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS', 'KIDNAPPING AND ABDUCTION OF OTHERS', 'DACOITY', 	'ROBBERY', 'BURGLARY', 'THEFT', 'AUTO THEFT' , 'OTHER THEFT', 'RIOTS', 'CRIMINAL BREACH OF TRUST', 'CHEATING', 'COUNTERFIETING', 'ARSON', 'HURT/GREVIOUS HURT', 'DOWRY DEATHS', 'ASSAULT ON WOMEN WITH INTENT TO OUTRAGE HER MODESTY', 'INSULT TO MODESTY OF WOMEN', 'CRUELTY BY HUSBAND OR HIS RELATIVES', 'IMPORTATION OF GIRLS FROM FOREIGN COUNTRIES' , 'CAUSING DEATH BY NEGLIGENCE']:
  for q in range(1):
    k = data_state.loc[:,['STATES',p]]
    kmean = KMeans(3)
    kmean.fit(k)
    cluster = k.copy()
    cluster['cluster_predict'] = kmean.fit_predict(k)
    plt.scatter(cluster['STATES'] ,cluster[p],c=cluster['cluster_predict'],cmap='brg')
    plt.xlabel('STATES')
    plt.ylabel(p)
    plt.show()

    fig = plt.figure(figsize = (10, 5))
 
    plt.bar(cluster.STATES, cluster[p], color ='green',
        width = 0.5)
 
    plt.xlabel("State")
    plt.ylabel(p)
    plt.title("Crime vs State")
    plt.show()   
```
